```python
%%writefile app.py
import streamlit as st
import google.generativeai as genai

if api_key:
    try:
        genai.configure(api_key=api_key)
        # Menggunakan 'gemini-1.5-flash-latest' seringkali lebih stabil untuk v1beta
        model = genai.GenerativeModel('gemini-1.5-flash-latest')

        # Test koneksi sederhana
        test_res = model.generate_content("hi")
    except Exception as e:
        st.error(f"Koneksi gagal: {e}")
        st.info("Mencoba beralih ke model gemini-pro...")
        model = genai.GenerativeModel('gemini-pro')

# --- KONFIGURASI HALAMAN ---
st.set_page_config(page_title="ASN Assistant", page_icon="🏛️")
st.title("🏛️ ASN Productivity Assistant")
st.markdown("---")

# --- SIDEBAR: API KEY ---
with st.sidebar:
    st.header("Konfigurasi")
    api_key = st.text_input("Masukkan Gemini API Key:", type="password")
    st.info("Dapatkan kunci di: aistudio.google.com")

if api_key:
    genai.configure(api_key=api_key)
    model = genai.GenerativeModel('gemini-1.5-flash')

    # --- MENU FITUR ---
    menu = ["Buat Surat", "Rangkum Regulasi", "Reminder Pekerjaan"]
    choice = st.selectbox("Pilih Fitur:", menu)

    if choice == "Buat Surat":
        st.subheader("✉️ Pembuatan Surat Otomatis")
        perihal = st.text_input("Perihal / Hal")
        tujuan = st.text_input("Tujuan (Yth.)")
        isi = st.text_area("Poin-poin isi surat")

        if st.button("Generate Draft"):
            prompt = f"Buat draft surat formal ASN. Perihal: {perihal}. Tujuan: {tujuan}. Isi: {isi}. Format: Tata Naskah Dinas, Bahasa Indonesia Formal."
            response = model.generate_content(prompt)
            st.code(response.text, language="markdown")

    elif choice == "Rangkum Regulasi":
        st.subheader("📜 Ringkasan Peraturan")
        teks = st.text_area("Tempel teks regulasi di sini...")
        if st.button("Proses Ringkasan"):
            prompt = f"Rangkum teks regulasi ini menjadi poin penting (Hak, Kewajiban, Sanksi): {teks}"
            response = model.generate_content(prompt)
            st.success(response.text)

    elif choice == "Reminder Pekerjaan":
        st.subheader("📅 Pengingat Tugas")
        tugas = st.text_input("Nama Tugas")
        tgl = st.date_input("Deadline")
        if st.button("Simpan"):
            st.balloons()
            st.write(f"✅ Tersimpan: {tugas} (Tenggat: {tgl})")
else:
    st.warning("Silakan masukkan API Key di sidebar untuk melanjutkan.")
```
